### Let's check the balance of the column 'fetal_health' in the dataset

```python
# Plot a countplot for fetal_health 
sns.countplot(x = df.fetal_health)
```

#### By observing the chart we say that the dataset is imbalanced.

```python
x = df.drop(columns = ['fetal_health', 'accelerations','uterine_contractions',
                       'mean_value_of_short_term_variability','mean_value_of_long_term_variability','histogram_width',
                       'histogram_max', 'histogram_number_of_peaks','histogram_mode', 'histogram_mean',
                       'histogram_median','histogram_tendency'])
```

#### Dropper every unwanted columns

```python
y = df.fetal_health
```

### Let's scale the x variable (independent)

```python
# For that we'll use StandardScaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
X = pd.DataFrame(x_scaled, columns = x.columns)
```

### Spliting the data into training and testing sets

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

### Let's build the model

```python
from sklearn.svm import SVC
model = SVC()
model.fit(X_train, y_train)
```

### Now the prediction

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
